```python
def calculate_cdf(lst, X):
    count = 0
    total_values = len(lst)

    for value in lst:
        if value <= X:
            count += 1

    cdf = count / total_values

    cdf = round(cdf, 3)

    return cdf

test_lst = [1, 2, 3]
test_X = 2

result = calculate_cdf(test_lst, test_X)
print(result)
```

```python
dice_lst = [1, 2, 3, 4, 5, 6]

dice_cum = []

for value in dice_lst:
    cdf_value = calculate_cdf(dice_lst, value)
    dice_cum.append(cdf_value)

print(dice_cum)
```

```python
import matplotlib.pyplot as plt

dice_lst = [1, 2, 3, 4, 5, 6]
dice_cum = [0.167, 0.333, 0.5, 0.667, 0.833, 1.0]

plt.figure(figsize=(8, 6))
plt.bar(dice_lst, dice_cum, width=0.5, align='center', alpha=0.7)

plt.title('Cumulative Distribution Function (CDF) of Fair Dice Roll')
plt.xlabel('Dice Value')
plt.ylabel('Cumulative Probability')
plt.xticks(dice_lst)
plt.ylim(0, 1)  # Set y-axis limit from 0 to 1
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

dice_lst = np.array([1, 2, 3, 4, 5, 6])

hist, bins = np.histogram(dice_lst, bins=6, range=(1, 7), density=True)

cumulative_prob = np.cumsum(hist)

plt.figure(figsize=(8, 6))
plt.scatter(bins[:-1], cumulative_prob, marker='o', color='blue', label='CDF')

plt.step(bins[:-1], cumulative_prob, where='post', color='blue')

plt.title('Cumulative Distribution Function (CDF) of Fair Dice Roll')
plt.xlabel('Dice Value')
plt.ylabel('Cumulative Probability')
plt.xticks(dice_lst)
plt.ylim(0, 1)  # Set y-axis limit from 0 to 1
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.legend()
plt.show()
```
